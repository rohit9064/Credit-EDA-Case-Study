```python
# Analsysis of variable due to imbalance
#CReating 2 different dataset for defaulters and other
Loan_defaulter=Loan_App[Loan_App['TARGET']==1]
Loan_other=Loan_App[Loan_App['TARGET']==0]
```

```python
#Now we will do analysis for categorical variables and numeric variables
#Univariate Analysis with respect to Defaulter and others for Categorical variables

CATEG_VAR=Loan_App.select_dtypes(include=['object']).columns
NUMERIC_VAR=Loan_App.select_dtypes(include=['int64','float64']).columns
```

```python
#Creating a function for plotting countplot

def countplot_function(col):
    plt.figure(figsize=(25,7))
    plt.subplot(221)
    sns.countplot(col,data=Loan_defaulter)
    plt.title('Count plot for Loan defaulter')
    plt.xticks(rotation=90)
    plt.subplot(222)
    sns.countplot(col,data=Loan_other)
    plt.title('Count plot for Other')
    plt.xticks(rotation=90)
    #plt.subplot(223)
    
    Percen = pd.DataFrame()
    Percen['Other']= ((Loan_other[col].value_counts())/len(Loan_other))
    Percen['Defaulter']= ((Loan_defaulter[col].value_counts())/len(Loan_defaulter))
    Percen.plot.bar()
    plt.title('Overall picture for percentage')
    plt.show()
    
```

```python
# Univariate Analysis for Categorical columns
#CATEG_VAR[0]

for i in range(0,len(CATEG_VAR)):
    print("COUNTPLOT for ", ' ', CATEG_VAR[i])
    countplot_function(CATEG_VAR[i])
    
```

```python
# Conclusion from above countplot

#1) NAME_CONTRACT_TYPE : The stats for both the categories ( defaulter and others ) looks identical ,
#    difficult to derive something,but we can say Revolving loans are less risky.
#2) CODE_GENDER: By looking the data it seems Female gets more loan compared to Male .
#                BUt percentage of male for default is more eventhough numbers are less.
#3) FLAG_OWN_CAR: Nothing much can be concluded but slight numbers are showing people with no cars are more defaulters.
#4) FLAG_OWN_REALTY: Nothing much can be concluded but slight numbers are showing people who have realty ==n are more defaulters.
#5) NAME_TYPE_SUITE: Nothing can be concluded.
#6) NAME_INCOME_TYPE: For working, defaulters are more than other and its opposite in case of pensioner
#7) NAME_EDUCATION_TYPE: For Higher education Defaulters are less.But loan taken for secondry education is more and more defaulters 
#                        are there for the same.
#8) NAME_FAMILY_STATUS: Married people have taken most loan but the percenrtage of defaulters are more in single/civil marriage
#9) NAME_HOUSING_TYPE: Defaulters ratio are more in case of with parents /rented apartmnet but loans taken are less.
#10)OCCUPATION_TYPE : BY seeing this laborers and sales staff have taken maximum loans but they are not trusted
#                     as number of defaulters are more. Some class which can be trusted are managers, accountants,High skilled
#                     tech staff
#11)WEEKDAY_APPR_PROCESS_START: Nothing much can be concluded.
#12)ORGANIZATION_TYPE:Nothing much can be concluded as lot of category are there .( please check bivariate analysis for conclusion)
#13)EMERGENCYSTATE_MODE: Nothing much can be concluded. 
```

```python
def plot_2variables(col, col1):
    fig = plt.figure(figsize=(25,20))

    ax1 = plt.subplot(221)
    Loan_App[col].value_counts().plot.pie(autopct = "%1.0f%%", ax=ax1)
    plt.title('Plotting data for the column: '+ col)


    ax2 = plt.subplot(222)
    sns.countplot(x=col, hue=col1, data=Loan_other_merged, ax = ax2)
    plt.xticks(rotation=90)
    plt.title('Plotting data for Others category')

    ax3 = plt.subplot(223)
    sns.countplot(x=col, hue=col1, data=Loan_defaulter_merged, ax = ax3)
    plt.xticks(rotation=90)
    plt.title('Plotting data for Defaulter')
    
    Percen = pd.DataFrame()
    Percen['Other']= ((Loan_other[col].value_counts())/len(Loan_other))
    Percen['Defaulter']= ((Loan_defaulter[col].value_counts())/len(Loan_defaulter))
    Percen.plot.bar()
    plt.title('Overall picture for percentage')



    fig.tight_layout() # Or equivalently,  "plt.tight_layout()"

    plt.show()
```
